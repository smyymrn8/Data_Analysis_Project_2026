```python
# Bağımsız değişkenler (X) ve hedef değişken (y) ayrılır.
X = df.drop("Attrition", axis=1)
y = df["Attrition"]
```

```python
from sklearn.model_selection import train_test_split

# Veriyi %80 eğitim, %20 test olacak şekilde bölüyoruz
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

print(f"Eğitim seti boyutu: {X_train.shape}")
print(f"Test seti boyutu: {X_test.shape}")
```

```python
# Veri setimizde (IBM Attrition) işten ayrılmayanlar (No), ayrılanlardan (Yes) çok daha fazla. Bu durum modelin "tembelleşmesine" ve herkese "ayrılmayacak" demesine sebep olur.
# Burada Smote yöntemi uygulanarak eğitim verisindeki sınıf dengesizliği giderilecek.

from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
# Smote sadece eğitim verisi üzerinde uygulanır.
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

print("Önce:", y_train.value_counts())
print("Sonra:", y_train_smote.value_counts())
```

```python
# Özelliklerin aynı ölçekte olması için standardizasyon uygulanır.
# bu sayede algoritmalar yüksek değerli değişkenlerde daha iyi performans gösterebilir.
# Burada ölçeklendirme sadece eğitim verisi üzerine fit edilir. 
# Test verisi model için hiç görülmemiş veri olmalıdır.
from sklearn.preprocessing import StandardScaler

# Scaler nesnesini oluşturuyoruz
scaler = StandardScaler()

# SADECE SMOTE uygulanmış EĞİTİM verisini fit_transform yapıyoruz
X_train_resampled = scaler.fit_transform(X_train_smote)

# TEST verisini ise sadece transform yapıyoruz (Eğitim setinin ortalamasına göre)
X_test_scaled = scaler.transform(X_test)

print("Eğitim ve Test verileri başarıyla ölçeklendirildi.")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Parametre aralıklarını belirliyoruz
param_grid_rf = {
    'n_estimators': [200, 250, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5]
}

# GridSearchCV tanımlıyoruz
grid_rf = GridSearchCV(RandomForestClassifier(random_state=42), param_grid_rf, cv=5, scoring='f1')

# Dengelenmiş ve Ölçeklenmiş eğitim verisiyle eğitiyoruz
grid_rf.fit(X_train_resampled, y_train_smote)

print("RF En iyi parametreler:", grid_rf.best_params_)

# Tahmin yaparken Ölçeklenmiş test verisini kullanıyoruz
y_pred_rf = grid_rf.best_estimator_.predict(X_test_scaled)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
import seaborn as sns
import matplotlib.pyplot as plt

# En iyi modelle tahmin yapıyoruz
best_rf = grid_rf.best_estimator_
y_pred_rf  = best_rf.predict(X_test_scaled)

# Confusion Matrix (Karışıklık Matrisi) görselleştirmesi
cm = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["No", "Yes"],
            yticklabels=["No", "Yes"])
plt.title("Random Forest Confusion Matrix")
plt.xlabel("Tahmin Edilen")
plt.ylabel("Gerçek Değer")
plt.show()
```

```python
# lojistik regresyon modeli için parametre aralığı.
from sklearn.linear_model import LogisticRegression

param_grid_lr = {
    "C": [0.01, 0.1, 1, 10],
    "solver": ["lbfgs"]
}

grid_lr = GridSearchCV(
    LogisticRegression(max_iter=1000),
    param_grid_lr,
    cv=5,
    scoring="f1",
    n_jobs=-1
)

# Modeli eğitiyoruz 
grid_lr.fit(X_train_resampled, y_train_smote)

# en iyi parametreleri görüyoruz. 
best_lr = grid_lr.best_estimator_

print("Best LR Params:", grid_lr.best_params_)
```

```python
# En iyi modelle tahmin yapıyoruz
best_lr = grid_lr.best_estimator_
y_pred_lr  = best_lr.predict(X_test_scaled)

# Confusion Matrix (Karışıklık Matrisi) görselleştirmesi
cm = confusion_matrix(y_test, y_pred_lr)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["No", "Yes"],
            yticklabels=["No", "Yes"])
plt.title("Lojistik Regresyon Confusion Matrix")
plt.xlabel("Tahmin Edilen")
plt.ylabel("Gerçek Değer")
plt.show()
```

```python
# xgboost modeli için parametre aralığı.
from xgboost import XGBClassifier

param_grid_xgb = {
    "n_estimators": [200, 250, 300],
    "max_depth": [3, 5, 7],
    "learning_rate": [0.01, 0.1]
}

grid_xgb = GridSearchCV(
    XGBClassifier(eval_metric='logloss', random_state=42),
    param_grid_xgb,
    cv=5,
    scoring="f1",
    n_jobs=-1
)

# modeli eğitiyoruz 
grid_xgb.fit(X_train_resampled, y_train_smote)

# en iyi parametreleri alıyoruz.
best_xgb = grid_xgb.best_estimator_
print("Best XGB Params:", grid_xgb.best_params_)
```

```python
# En iyi modelle tahmin yapıyoruz
best_xgb = grid_xgb.best_estimator_
y_pred_xgb  = best_xgb.predict(X_test_scaled)

# Confusion Matrix (Karışıklık Matrisi) görselleştirmesi
cm = confusion_matrix(y_test, y_pred_xgb)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["No", "Yes"],
            yticklabels=["No", "Yes"])
plt.title("XGBoost İçin Confusion Matrix")
plt.xlabel("Tahmin Edilen")
plt.ylabel("Gerçek Değer")
plt.show()
```

```python
# Model karşılaştırılması için temel metrikler hesalanacak ve tablo halinde sunulacak.
# Tahminleri ölçeklendirilmiş test verisi ile yeniden yapıyoruz
# Bu adım 'feature names' uyarısını ve performans düşüklüğünü çözecektir.
y_pred_rf = best_rf.predict(X_test_scaled)
y_pred_lr = best_lr.predict(X_test_scaled)
y_pred_xgb = best_xgb.predict(X_test_scaled)

# 2. Sonuç tablosunu yeniden oluşturuyoruz
results = pd.DataFrame({
    "Model": ["Random Forest", "Logistic Regression", "XGBoost"],
    "Accuracy": [
        accuracy_score(y_test, y_pred_rf),
        accuracy_score(y_test, y_pred_lr),
        accuracy_score(y_test, y_pred_xgb)
    ],
    "Precision": [
        precision_score(y_test, y_pred_rf, zero_division=0),
        precision_score(y_test, y_pred_lr, zero_division=0),
        precision_score(y_test, y_pred_xgb, zero_division=0)
    ],
    "Recall": [
        recall_score(y_test, y_pred_rf),
        recall_score(y_test, y_pred_lr),
        recall_score(y_test, y_pred_xgb)
    ],
    "F1 Score": [
        f1_score(y_test, y_pred_rf),
        f1_score(y_test, y_pred_lr),
        f1_score(y_test, y_pred_xgb)
    ]
})

# Tabloyu ekranda daha şık görmek için:
print(results.to_string(index=False))
```

```python
import pandas as pd
import numpy as np
import random
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score

# 1. 30 Tane Benzersiz Seed (Tohum) Üretilmesi
# Benzer veri setlerinin oluşmasını engellemek için 0-10000 arası karıştırılmış liste kullanıyoruz.
seed_pool = list(range(10001))
random.shuffle(seed_pool)
seeds = seed_pool[:30] # İlk 30 benzersiz sayıyı alıyoruz.

# Sonuçları saklamak için listeler oluşturalım
results_list = []

print("30 farklı veri seti simülasyonu başlatılıyor...")

# 2. Döngü İçerisinde 30 Farklı Veri Bölme ve Test İşlemi
for i, s in enumerate(seeds):
    # Veri setini her döngüde farklı bir seed ile bölüyoruz.
    X_train_iter, X_test_iter, y_train_iter, y_test_iter = train_test_split(
        X, y, test_size=0.20, random_state=s
    )
    
    # Not: Hoca eğitim ve test kullanarak yapmamızı istediği için ölçeklendirme 
    # her seferinde eğitim verisine göre yeniden yapılmalıdır.
    from sklearn.preprocessing import StandardScaler
    scaler_iter = StandardScaler()
    X_train_scaled = scaler_iter.fit_transform(X_train_iter)
    X_test_scaled = scaler_iter.transform(X_test_iter)
    
    # SMOTE ile dengeleme (Dengesizliği gidermek için şarttır)[cite: 1]
    from imblearn.over_sampling import SMOTE
    smote_iter = SMOTE(random_state=42)
    X_train_res, y_train_res = smote_iter.fit_resample(X_train_scaled, y_train_iter)

    # Modellerin bu özel veri setinde eğitilmesi ve tahmini
    # Random Forest
    best_rf.fit(X_train_res, y_train_res)
    rf_f1 = f1_score(y_test_iter, best_rf.predict(X_test_scaled))
    
    # Logistic Regression
    best_lr.fit(X_train_res, y_train_res)
    lr_f1 = f1_score(y_test_iter, best_lr.predict(X_test_scaled))
    
    # XGBoost
    best_xgb.fit(X_train_res, y_train_res)
    xgb_f1 = f1_score(y_test_iter, best_xgb.predict(X_test_scaled))
    
    # Her bir iterasyonun sonucunu kaydediyoruz[cite: 1].
    results_list.append({
        "Iteration": i + 1,
        "Seed": s,
        "Random Forest F1": rf_f1,
        "Logistic Regression F1": lr_f1,
        "XGBoost F1": xgb_f1
    })

# 3. Sonuçların Tablo Halinde Gösterilmesi[cite: 1]
comparison_df = pd.DataFrame(results_list)

pd.set_option('display.max_rows', None) 

print("--- 30 Farklı Seed İçin Model Performans Sonuçları ---")

display(comparison_df) 

# İstatistiksel özet (Describe) - ortalama ve sapma değerleri için
print("\n--- Modellerin 30 Deneme Üzerindeki İstatistiksel Özeti ---")
print(comparison_df[['Random Forest F1', 'Logistic Regression F1', 'XGBoost F1']].describe())
```

```python
# 1. İstikrar (Stability) Kontrolü: Modellerin başarısı sadece tek bir veri ayrımına mı (seed) bağlı, 
# yoksa farklı veri senaryolarında da tutarlı mı kalıyor? Bunu kutuların darlığından anlıyoruz.
# Bu 30 farklı durumun dağılımını tesadüf mü yoksa gerçekten best model mi olduğunu görmek için görselleştireceğiz.
plot_df = comparison_df.melt(id_vars=['Iteration', 'Seed'], 
                             value_vars=['Random Forest F1', 'Logistic Regression F1', 'XGBoost F1'],
                             var_name='Model', value_name='F1 Score')

plt.figure(figsize=(12, 6))
sns.boxplot(
    x='Model', 
    y='F1 Score', 
    data=plot_df, 
    hue='Model', # Renklendirmeyi modele göre yapıyoruz
    palette='Set3', 
    legend=False # Gereksiz kutucuğu kapatıyoruz
)
sns.stripplot(x='Model', y='F1 Score', data=plot_df, color='black', alpha=0.3) # Bireysel noktaları da görelim
plt.title('30 Farklı Veri Bölümü Üzerinde Model Performanslarının Karşılaştırılması')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

*Kutu ne kadar darsa modelin o kadar istikrarlı olduğu sonucuna ulaşmış olacağız.
*Ortadaki medyan çizgisi ise ilgili modelin genel performansını gösteriyor.

Lojistik Regresyon için: Boxplotun nispeten daha dar olması, veriyi 30 farklı şekilde bölsek de modelin benzer sonuçlar ürettiğini, yani kararlı (stable) bir performans sergilediğini kanıtlıyor. Bu da modelin gerçekten öğrendiğini ve tesadüfi sonuçlar doğurmadığı sonuçlarını çıkarmamızı sağlar.

Random Forest ve XGBoost için ise boxplotlar daha geniş görünüyor. Bu durum, bu modellerin veri bölünmesine karşı biraz daha hassas olduğunu ve bazı seed'lerde performans dalgalanmaları yaşadığını gösterir.

Aykırı Değerler ve Dağılım: Her üç modelde de kutuların çok dışında kalan aşırı uç noktalar (outliers) bulunmuyor. Bu, modellerin genel olarak veri setine uyum sağladığını ve "overfitting" (ezberleme) riskinin kontrol altında olduğunu işaret eder.

Sonuç: 30 farklı rastgele veri bölümü üzerinde yaptığımız testler sonucunda; modellerin performansının tesadüfi bir veri ayrımına bağlı olmadığını, Logistic Regression modelinin özellikle istikrar ve genel başarı açısından diğer algoritmalara göre daha güvenilir bir performans sergilediğini Boxplot üzerinden doğruladık.

```python
from scipy.stats import wilcoxon

# Hoca 2'şerli kıyaslama ve 6 durum (A-B, B-A dahil edilirse veya ek metriklerle) dediği için:
# Karşılaştırmaları 30'luk skor listeleri üzerinden yapıyoruz

# 1. RF vs LR
stat_rf_lr, p_rf_lr = wilcoxon(comparison_df["Random Forest F1"], comparison_df["Logistic Regression F1"])
print(f"RF vs LR p-value: {p_rf_lr:.10f}")

# 2. RF vs XGB
stat_rf_xgb, p_rf_xgb = wilcoxon(comparison_df["Random Forest F1"], comparison_df["XGBoost F1"])
print(f"RF vs XGB p-value: {p_rf_xgb:.10f}")

# 3. LR vs XGB
stat_lr_xgb, p_lr_xgb = wilcoxon(comparison_df["Logistic Regression F1"], comparison_df["XGBoost F1"])
print(f"LR vs XGB p-value: {p_lr_xgb:.10f}")

# YORUM: P < 0.05 ise bu iki model arasındaki fark tesadüf değildir.
```

Yapılan Wilcoxon testleri sonucunda, tüm model çiftleri için elde edilen p-değerleri 0.05 kritik değerinin altında kalmıştır. Bu durum, 30 farklı veri seti üzerinden elde edilen F1 skorları arasındaki farkların istatistiksel olarak anlamlı olduğunu ve modellerin performans sıralamasının tesadüfi bir veri bölünmesine bağlı olmadığını kanıtlamaktadır. 

Özellikle Logistic Regression modelinin sergilediği yüksek başarı, şans eseri değil, verideki örüntüleri tutarlı bir şekilde öğrenmesinin sonucudur.

```python
# Genel bir kıyaslama için derste bahsedilen Friedman Testi'ni uygulayacağız.
# Friedman testi, üç veya daha fazla bağımsız grubun medyanlarını karşılaştırmak için kullanılan non-parametrik bir testtir.

from scipy.stats import friedmanchisquare
stat, p = friedmanchisquare(comparison_df["Random Forest F1"], 
                            comparison_df["Logistic Regression F1"], 
                            comparison_df["XGBoost F1"])
print(f"Friedman Testi p-value: {p:.10f}")
```

H0: "3 modelin performansları arasında istatistiksel olarak hiçbir fark yoktur, hepsi aynıdır." -> Bu hipotez red durumuna düştü

H1: "Modellerden en az biri, diğerlerinden istatistiksel olarak anlamlı derecede farklı ve daha başarılıdır." -> Bu hipotez kabul edilebilir durumda.

Uyguladığımız Friedman Testi sonucunda elde ettiğimiz 0.0000957053 p-değeri, kullandığımız üç algoritmanın (Random Forest, Logistic Regression ve XGBoost) performansları arasında istatistiksel olarak çok güçlü ve anlamlı bir fark olduğunu kanıtlamaktadır. Bu sonuç, modellerden birinin (Boxplot grafiğinde gördüğümüz üzere Logistic Regression) diğerlerine göre şans eseri değil, yapısal olarak daha üstün bir performans sergilediğini bilimsel olarak doğrular.
