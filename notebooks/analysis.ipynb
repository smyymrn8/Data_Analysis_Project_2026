```python
# Model için anlamlı olmayan sütunlar kaldırılır
df.drop(["EmployeeCount", "EmployeeNumber", "Over18", "StandardHours"], axis=1, inplace=True)
```

```python
# Attrition sütunu hedef değişkendir
# Yes/No değerleri sayısal hale getirilir (1/0)
df["Attrition"] = df["Attrition"].map({"Yes": 1, "No": 0})
```

```python
# Makine öğrenmesi algoritmaları string veri ile çalışamaz.
# Bu nedenle kategorik veriler sayısal değerlere dönüştürülecek.
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in df.select_dtypes(include="object").columns:
    df[col] = le.fit_transform(df[col])
```

```python
sns.countplot(x="Attrition", data=df)
plt.title("Attrition Dağılımı")
plt.show()
```

```python
# Bağımsız değişkenler (X) ve hedef değişken (y) ayrılır.
X = df.drop("Attrition", axis=1)
y = df["Attrition"]
```

```python
# Özelliklerin aynı ölçekte olması için standardizasyon uygulanır.
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

# Veriyi %80 eğitim, %20 test olacak şekilde bölüyoruz
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

print(f"Eğitim seti boyutu: {X_train.shape}")
print(f"Test seti boyutu: {X_test.shape}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Denenecek parametre aralıklarını belirliyoruz
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

# GridSearchCV tanımlıyoruz
grid_rf = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=5, scoring='f1')

# Modeli eğitiyoruz
grid_rf.fit(X_train, y_train)

# En iyi parametreleri görüyoruz
print("En iyi parametreler:", grid_rf.best_params_)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
import seaborn as sns
import matplotlib.pyplot as plt

# En iyi modelle tahmin yapıyoruz
best_rf = grid_rf.best_estimator_
y_pred_rf  = best_rf.predict(X_test)

# Confusion Matrix (Karışıklık Matrisi) görselleştirmesi
cm = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["No", "Yes"],
            yticklabels=["No", "Yes"])
plt.title("Random Forest Confusion Matrix")
plt.xlabel("Tahmin Edilen")
plt.ylabel("Gerçek Değer")
plt.show()
```

```python
# lojistik regresyon modeli için parametre aralığı.
from sklearn.linear_model import LogisticRegression

param_grid_lr = {
    "C": [0.01, 0.1, 1, 10],
    "solver": ["lbfgs"]
}

grid_lr = GridSearchCV(
    LogisticRegression(max_iter=1000),
    param_grid_lr,
    cv=5,
    scoring="f1",
    n_jobs=-1
)

# modeli eğitiyoruz.
grid_lr.fit(X_train, y_train)

# en iyi parametreleri görüyoruz. 
best_lr = grid_lr.best_estimator_

print("Best LR Params:", grid_lr.best_params_)
```

```python
# En iyi modelle tahmin yapıyoruz
best_lr = grid_lr.best_estimator_
y_pred_lr  = best_lr.predict(X_test)

# Confusion Matrix (Karışıklık Matrisi) görselleştirmesi
cm = confusion_matrix(y_test, y_pred_lr )
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["No", "Yes"],
            yticklabels=["No", "Yes"])
plt.title("Lojistik Regresyon Confusion Matrix")
plt.xlabel("Tahmin Edilen")
plt.ylabel("Gerçek Değer")
plt.show()
```

```python
# xgboost modeli için parametre aralığı.
from xgboost import XGBClassifier

param_grid_xgb = {
    "n_estimators": [50, 100],
    "max_depth": [3, 5, 7],
    "learning_rate": [0.01, 0.1]
}

grid_xgb = GridSearchCV(
    XGBClassifier(eval_metric='logloss'),
    param_grid_xgb,
    cv=5,
    scoring="f1",
    n_jobs=-1
)

grid_xgb.fit(X_train, y_train)

best_xgb = grid_xgb.best_estimator_

print("Best XGB Params:", grid_xgb.best_params_)
```

```python
# En iyi modelle tahmin yapıyoruz
best_xgb = grid_xgb.best_estimator_
y_pred_xgb  = best_xgb.predict(X_test)

# Confusion Matrix (Karışıklık Matrisi) görselleştirmesi
cm = confusion_matrix(y_test, y_pred_xgb)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["No", "Yes"],
            yticklabels=["No", "Yes"])
plt.title("XGBoost İçin Confusion Matrix")
plt.xlabel("Tahmin Edilen")
plt.ylabel("Gerçek Değer")
plt.show()
```
